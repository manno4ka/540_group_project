In this notebook. We are analyzing the effect of the job level on the attrition rate.

```python
import numpy as np
import pandas as pd
import os
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
path = os.path.join("..","WA_Fn-UseC_-HR-Employee-Attrition.csv")
```

```python
df = pd.read_csv(path)
```

Doing some grouping, let's analyze the job level per department

```python
sns.displot(data=df,x="JobLevel",hue="Attrition")
plt.show()
```

```python
df[["Department","Attrition","JobLevel"]].groupby(["Department","JobLevel","Attrition"]).size().groupby(level=[0,1]).apply(lambda x: x/x.sum())
```

```python
df[["EducationField","Attrition","JobLevel","Department"]].groupby(["Department","EducationField","JobLevel","Attrition"]).size().groupby(level=[0,1,2]).apply(lambda x: x/x.sum())
```
