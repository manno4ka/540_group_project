```python
import numpy as np
import pandas as pd
import os
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
path = os.path.join("..","WA_Fn-UseC_-HR-Employee-Attrition.csv")
```

```python
df = pd.read_csv(path)
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe().T
```

#### Dropped below columns because it's only has one value for all rows.

```python
df.drop(columns=['EmployeeCount', 'Over18', 'StandardHours'], inplace=True)
```

```python
dfColumns = []
for i in df.columns:
    dfColumns.append([i, df[i].nunique(), df[i].drop_duplicates().values])
pd.DataFrame(dfColumns, columns = ['Features', 'Unique Number', 'Values'])
```

#### Number 1 is Yes, means employee moves out of company and number 0 is No, means stay.

```python
df['Attrition'] = np.where(df['Attrition'] == 'Yes', 1, 0)
```

```python
df['Attrition'].value_counts()/df.shape[0]*100
```

```python
dfCol = []
for i in df.columns:
    dfCol.append([i, df[i].nunique(), df[i].drop_duplicates().values])
pd.DataFrame(dfCol, columns = ['Columns', 'Unique Number', 'Values'])
```

```python
# Correlation Plot
f, ax = plt.subplots(figsize=(20, 15))
corr = df.corr()
hm = sns.heatmap(round(corr,2), annot=True, ax=ax, cmap="coolwarm",fmt='.2f', linewidths=.05)
f.subplots_adjust(top=0.95)
t= f.suptitle('HR-Employee-Attrition - Correlation Heatmap', fontsize=16)
```

#### We can see from the correlation plot that Attrition does not have strong correlation with any other variables.
#### We can also see from the correlation plot that only few variables have relativley strong correaltion. 
#### Job Level and Monthly Income has highet correlation of 0.95, which makes sense. 

#### Some other correlated varibles
#### TotalWorkingYears and Monthly Income (0.78)
#### Age and TotalWorkingYears (0.68)
#### YearsAtComapny and YearsWithCurrManager (0.77)
#### YearsInCurrentRole and YearsWithCurrManager (0.71)
#### YearsInCurrentRole and YearsAtCompany (0.76)
